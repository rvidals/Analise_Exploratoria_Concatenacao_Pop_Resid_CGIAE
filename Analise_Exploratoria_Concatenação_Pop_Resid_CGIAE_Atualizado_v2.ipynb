# Código para concatenar vários csv e ordenar a partir do Sexo e o código do Município


```python
#Importando a biblioteca pandas
import pandas as pd
from google.colab import files
```

```python
#Importar os arquivos csv e salvando na variável pop

pop = files.upload()
```

```python
#Caminho onde serão salvo os arquivos
path = '/content/'
```

```python
li_M = [] # os valores são atribuidos à uma lista
li_F = [] # os valores são atribuidos à uma lista


# o método key foi usado para chamar os  nomes dos arquivos presentes no google colab
for name in pop.keys():

  #Essa parte é para pegar a faixa etária e renomear a coluna sexo
  df_fEtaria  = pd.read_csv(path+name, encoding='latin1', header=None)
  # fEtaria = df_fEtaria.iat[3,0]
  fEtaria = df_fEtaria.iat[2,0].replace(';','')
  print(fEtaria)
  fEtaria = fEtaria.split(':')[1].replace('  ',' ').replace(' ','_')
  
  #print(name,': ',fEtaria) #visualizar a execução da extração

  #Essa parte é para pegar o csv que gerará a nava tabela
  # o csv vem com informações além das tabulaeres 
  # Para fazer essa limpeza foi utilizados as flegues skiprows, skipfooter e header
  df = pd.read_csv(path+name, skiprows=2, encoding='latin1', header=2, sep=';')
  df2 = df.rename(columns={"Masculino": "Masculino"+fEtaria, "Feminino": "Feminino"+fEtaria})
  
  #Lista onde será salvo as colunas da tabela para concatenar 
  li_M.append(df2["Masculino"+fEtaria]) # atribuindo as colunas renomeadas à lista
  li_F.append(df2["Feminino"+fEtaria]) # atribuindo as colunas renomeadas à lista
```

```python
li_IBGE = [] # o Código do IBGE e nome do município será atribuido a esta lista

li_IBGE.append(df["Município"].str.split(' ',
                                         1,
                                         expand=True).rename(columns = {0: 'COD_MUN',
                                                                       1:'NM_MUN'})) 

#Estou inserindo os dados do ibge na lista masculina
li_IBGE_M = li_IBGE + li_M 
#Estou inserindo os dados do ibge na lista masculina
li_IBGE_F = li_IBGE + li_F

#Concatenando os valores em um DataFrame
concat_M = pd.concat(li_IBGE_M , axis=1) # concatenando pelo eixo y = 1
concat_F = pd.concat(li_IBGE_F, axis=1) #Vizualização
```
